# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model

# Declare a Base using `automap_base()`
Base = automap_base()

# reflect the tables

# Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found

# Print all of the classes mapped to the Base
Base.classes.keys()
```

```python
# Save references to each table

# Assign the measurement class to a variable called `Measurement`
Measurement = Base.classes.measurement

# Assign the station class to a variable called `Station`
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB

# Create a session
session = Session(engine)
```

```python
# Display the Measurement class row's columns and data in dictionary format
first_row = session.query(Measurement).first()
first_row.__dict__
```

```python
# Display the Station class row's columns and data in dictionary format
first_row = session.query(Station).first()
first_row.__dict__
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```

```python
# Create a query that will return the daily temperatures
# for all historic data matching a specific month

def monthly_normals(date):
    """daily temperature for month specified.
    
    Args:
        date (str): A date string in the format '%m'
        
    Returns:
        A list of tuples containing the daily temp.
    
    """
    
    sel = [Measurement.tobs]
    return session.query(*sel).filter(func.strftime("%m", Measurement.date) == date)\
    .limit(1500).all()

# test function for February
monthly_normals("02")
```

```python
# Using function monthly_normals(), get tenmperatures for June and December
june_temps = monthly_normals('06')
dec_temps = monthly_normals('12')

print(len(june_temps))
print(len(dec_temps))
```

```python
june_temps = np.ravel(june_temps)
dec_temps = np.ravel(dec_temps)
#print(june_temps)
```

```python
temp_df = pd.DataFrame({"June": june_temps,
                       "Dec": dec_temps})
temp_df.head()
```

```python
temp_df.describe()
```

```python
temp_df =  temp_df.sample(n=150)
temp_df.describe()
```

```python
stats.shapiro(june_temps)
```

```python
stats.shapiro(dec_temps)
```

Which t-test to choose? The data is coming from same 9 Hawaii stations(same stations & same locations of the stations) for two different periods of time). These are same subjects (dependent) which is why I choose "paired" t-test. The distraibution of each sample is also "normally" distributed as shown below.

```python
#stats.ttest_ind(june_temps,dec_temps)
```

```python
# Perform paired t-test on two samples
stats.ttest_rel(june_temps,dec_temps)
```

```python
_,p_value = stats.ttest_rel(june_temps,dec_temps)
```

```python
p_value
```

```python
if p_value < 0.05:
    print("We reject null hypothesis!")
    print ("Null Hypothesis was - there is NO statistical difference in means of two populations")
    print("Conclusion - the means of the two samples are significantly different")
else:
    print("We accept null hypothesis!")
    print ("Null Hypothesis was - there is NO statistical difference in means of two populations")
    print("Conclusion - the means of the two samples are NOT significantly different")
```

```python
def plot_distribution(input_data):
    plt.figure()
    ax = sns.distplot(input_data)
    plt.axvline(np.mean(input_data), color='k', linestyle='dashed', linewidth=5)
    _, max_ = plt.ylim()
    plt.text(
            input_data.mean() + input_data.mean() /10,
            max_ -  max_ / 10,
            "Mean: {:.2f}".format(input_data.mean()),
    )
    return plt.figure
```

```python
plot_distribution(temp_df['June'])
```

```python
plot_distribution(temp_df['Dec'])
```

FINAL CONCLUSION: The two samples of June and Dec temps came from same 9 stations making the measurement, and show close to normal distribution so used paired t-test to compare the means. The p-value is significantly powered than assumed 0.05 level of significance. Hence rejected null hypothesis and concluded sample means are statistically different. 

The mean temp of Dec is 71 deg F vs June is 75 deg F. This also does not differe when comparing 1500 samples from data and random 150 samples from same dataset.
